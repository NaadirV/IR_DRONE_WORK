# Drone Plots

For this notebook, we look at the factors tha influence drone flight time. 


## Flight Time plot

These are the plots showing the dependancies of the flight time equation

$$
FT = BC \times \frac{BD}{AAD} \times 60
$$
where:

$FT$ = Flight time

$BC$ = Battery Capacity

$BD$ = Baterry Discharge

$AAD$ = Average Amp Draw

It is given below, that the standard for $BD$ is 80% due to constant power leakage from the battery even when not in use. Due to the fact that $BC$ generally ranges from 2200mAh to 4480mAh for drones, an average of 3300mAh was chosen for this plot.


```python
#Initializations
import numpy as np
import matplotlib.pyplot as plt

#Axis lables
plt.xlabel("Average Amp Draw [A]")
plt.ylabel("Flight Time [min]")
plt.grid()

#The function to be plotted
Average_Amp_Draw = np.linspace(1,30,1000)
Flight_time = (3.8*(0.8/Average_Amp_Draw))*60

#Plotting the function
plt.plot(Average_Amp_Draw,Flight_time)
plt.show()
```

The plot below indicates the relationship between flight time and battery capacity. An average of 20amps will be used to set the AAD to be a constant

```python
#Initializations
import numpy as np
import matplotlib.pyplot as plt

#Axis Labels
plt.xlabel("Battery Capacity [Ah]")
plt.ylabel("Flight Time [min]")
plt.grid()

#The function to be plotted
BC = np.linspace(0,10,1000)
Flight_time = (BC*(0.8/20))*60

#Plotting the results
plt.plot(BC,Flight_time)
plt.show()
```

Whilst the plot appears to be linear, it actual tends to curve and saturate at a maximum due to the fact that with an increased battery capacity, there will be a physical increase in the size of the battery, hence more weight added to the drone which in turn leads to a lower flight time.

```python
#Initializations
def f(BC, AAD):
    return (BC*(0.8/AAD))*60

BC = np.linspace(0, 10, 1000)
AAD = np.linspace(1, 10, 1000)

X, Y = np.meshgrid(BC, AAD)
Z = f(X, Y)

#Arranging the figure with lables
fig = plt.figure(figsize=(10,10))
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 400,
                cmap='viridis')
ax.set_xlabel('Battery Capacity [Ah]')
ax.set_ylabel('Average Amp Draw [A]')
ax.set_zlabel('Flight Time [min]');

#Basically showing the plot
ax.view_init(20, 40)
```

The plot above indicates the 3 dimensional relationship between Battery Capacity, Average Amp Draw and Flight Time.
